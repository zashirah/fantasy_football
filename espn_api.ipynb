# Get teams with players

```python
def get_player(player, team_id):
    player_d = {}
    player_id = str(player.playerId)
    player_d['player_id'] = player_id
    player_d['team_id'] = team_id
    player_d['player_name'] = player.name
    player_d['posRank'] = player.posRank
    player_d['eligible_slots'] = player.eligibleSlots
    player_d['lineupSlot'] = player.lineupSlot
    player_d['acquisitionType'] = player.acquisitionType
    player_d['proTeam'] = player.proTeam
    # player_d['schedule'] = player.schedule
    player_d['onTeamId'] = player.onTeamId
    player_d['position'] = player.position
    player_d['injuryStatus'] = player.injuryStatus
    player_d['injured'] = player.injured
    player_d['total_points'] = player.total_points
    player_d['avg_points'] = player.avg_points
    player_d['projected_total_points'] = player.projected_total_points
    player_d['projected_avg_points'] = player.projected_avg_points
    player_d['percent_owned'] = player.percent_owned
    player_d['percent_started'] = player.percent_started
    player_d['stats'] = player.stats

    print(player_d)

    with open(f'data/espn_api/players/players_{player_id}_{file_timestamp}.json', 'w') as f:
        json.dump(player_d, f)
    
```

```python
def get_team_players(team):
    [get_player(player, str(team.team_id)) for player in team.roster]
```

```python
def get_team(team):
    team_d = {}
    team_id = str(team.team_id)
    team_d['team_id'] = team_id
    team_d['team_abbrev'] = team.team_abbrev
    team_d['team_name'] = team.team_name
    team_d['division_id'] = team.division_id
    team_d['division_name'] = team.division_name
    team_d['wins'] = team.wins
    team_d['losses'] = team.losses
    team_d['ties'] = team.ties
    team_d['points_for'] = team.points_for
    team_d['points_against'] = team.points_against
    team_d['waiver_rank'] = team.waiver_rank
    team_d['acquisitions'] = team.acquisitions
    team_d['acquisition_budget_spent'] = team.acquisition_budget_spent
    team_d['drops'] = team.drops
    team_d['trades'] = team.trades
    team_d['owners'] = team.owners
    team_d['streak_type'] = team.streak_type
    team_d['streak_length'] = team.streak_length
    team_d['standing'] = team.standing
    team_d['final_standing'] = team.final_standing
    team_d['draft_projected_rank'] = team.draft_projected_rank
    team_d['playoff_pct'] = team.playoff_pct
    team_d['logo_url'] = team.logo_url
    team_d['created_ts'] = file_timestamp

    with open(f'data/espn_api/teams/teams_{team_id}_{file_timestamp}.json', 'w') as f:
        json.dump(team_d, f)
        
```

```python
for team in league.teams:
    get_team(team)
    get_team_players(team)
```

# get free agents

```python
def get_fa_list():
    position_list = ['QB', 'RB', 'WR', 'TE', 'D/ST', 'K', 'FLEX']
    fa_list = []
    for position in position_list:    
        fa_list.extend(league.free_agents(size=50, position=position))

    return fa_list
```

```python
def get_fas():
    return [get_player(player, 'FA') for player in get_fa_list()]

```

```python
get_fas()
```

# get box scores

```python
def get_box_player(player, week, team):
    player_d = {}
    player_d['player_id'] = str(player.playerId)
    player_d['week'] = week
    player_d['team'] = team
    player_d['slot_position'] = player.slot_position
    player_d['points'] = player.points
    player_d['projected_points'] = player.projected_points
    player_d['slot_position'] = player.slot_position
    player_d['pro_opponent'] = player.pro_opponent
    player_d['pro_pos_rank'] = player.pro_pos_rank
    player_d['game_played'] = player.game_played
    player_d['on_bye_week'] = player.on_bye_week
    player_d['active_status'] = player.active_status

    with open(f'data/espn_api/box_players/box_players_{player.playerId}_{week}.json', 'w') as f:
        json.dump(player_d, f)


```

```python
def get_box_score(box_score, week):
    box_score_d = {}
    home_team = str(box_score.home_team)
    away_team = str(box_score.away_team)
    box_score_d['week'] = week
    box_score_d['home_team'] = home_team
    box_score_d['home_score'] = box_score.home_score
    box_score_d['home_projected'] = box_score.home_projected
    box_score_d['away_team'] = away_team
    box_score_d['away_score'] = box_score.away_score
    box_score_d['away_projected'] = box_score.away_projected
    box_score_d['is_playoff'] = box_score.is_playoff
    box_score_d['matchup_type'] = box_score.matchup_type

    # [get_box_player(player, week, home_team) for player in box_score.home_lineup]
    # [get_box_player(player, week, away_team) for player in box_score.away_lineup]

    with open(f'data/espn_api/box_scores/box_scores_{week}_{home_team}_{away_team}.json', 'w') as f:
        json.dump(box_score_d, f)
```

```python
def get_week_box_scores(week: int):
    for box_score in league.box_scores(week):
        get_box_score(box_score, week)
        [get_box_player(player, week, str(box_score.home_team)) for player in box_score.home_lineup]
        [get_box_player(player, week, str(box_score.away_team)) for player in box_score.away_lineup]
```

```python
for week in range(1,15):
    get_week_box_scores(week)

```
